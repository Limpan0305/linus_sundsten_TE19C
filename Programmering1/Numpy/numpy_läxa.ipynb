```python
#3
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,11)
print(x)
print(f"x={x}")
def f(k,x,m):
  y=k*x+m
  return y
print(f"y={f(3,x,1)}")
plt.plot(x,f(3,x,1))
plt.title("En rät linje")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#4
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-2,2)
print(x)
print(f"x={x}")
def f(k,x,m):
  y=k*x+m
  return y
print(f"y={f(x,x,-1)}")
plt.plot(x,f(x,x,-1))
plt.title("En kurva")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#5
import matplotlib.pyplot as plt
import numpy as np
print(f"x={x}")
x=np.linspace(0,6)
def f(a,b,c,t):
  y=(a*t**2)+(b*t)+c
  return y
print(f"y={round(max(f(-4,24,1,x)))} Är maximala höjden den når")
plt.plot(x,f(-4,24,1,x))#sätter x =0
plt.title("En kurva")
plt.xlabel("Tid")
plt.ylabel("Höjd")
plt.show()
```
