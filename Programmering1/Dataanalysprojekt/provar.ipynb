```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("Gender_Data.csv") #Läser in filen med hjälp av pandas
smittade = df["Total_Cases"]
färg = ["blue", "pink"]
döda = df["Total_Deaths"]


def cirkel(x,y):
    fig, axs = plt.subplots(1,2, figsize=(12,4), sharey=True)
    axs[0].pie(x, labels=y, autopct='%1.1f%%', colors=färg)
    axs[0].set_xlabel('Ålder', fontsize=10)
    axs[0].set_ylabel('Nåogt', fontsize=10)
    axs[1].plot(x,y)
    fig.subtitle('Antal döda bland könen', fontsize=20)
    axs[1].set_xlabel('common_xlabel', fontsize=10)
    axs[1].set_xlabel('common_xlabel', fontsize=10)
cirkel(döda,döda)
```
