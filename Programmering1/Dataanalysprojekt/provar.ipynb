```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("Gender_Data.csv") #Läser in filen med hjälp av pandas
smittade = df["Total_Cases"]
färg = ["blue", "pink"]
döda = df["Total_Deaths"]

x = ("Män", "Kvinnor")
def cirkel(x,y):
    plt.pie(x, labels=y, autopct='%1.1f%%', colors=färg)
    plt.axis('equal')
    plt.title("Antal smittade bland könen")
    plt.show()
cirkel(smittade,smittade)
plt.subplot(2,2,1)
plt.plot(x,smittade)

def cirkel(x,y):
    plt.pie(x, labels=y, autopct='%1.1f%%', colors=färg) 
    plt.axis('equal')
    plt.title("Antal döda bland könen")
    plt.show()
cirkel(döda,döda)
plt.subplot(2,2,2)
plt.plot(x,döda)
```
