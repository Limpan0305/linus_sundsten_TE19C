```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("Gender_Data.csv") #Läser in filen med hjälp av pandas
smittade = df["Total_Cases"]
färger = ["blue", "pink"]
döda = df["Total_Deaths"]
x=("Män", "Kvinnor")

def grafer(x,y):
  fig, axs = plt.subplots(1, 2, figsize=(16, 6))
  axs[0].bar(x, y, width=0.8, color=färger, align="center")
  axs[0].set_xlabel("Kön")
  axs[0].set_ylabel("Totala fall")
  my_labels = 'Manliga fall', 'Kvinnliga fall'
  axs[1].pie(y, labels = my_labels, autopct = '%1.1f%%', startangle = 2, colors=färger) 
  fig.suptitle('Manliga & Kvinnliga totala fall och dödsfall')
grafer(x,döda)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("National_Daily_Deaths.csv") #Läser in filen med hjälp av pandas
döda = df["National_Daily_Deaths"]
färger = ["blue", "pink"]
x=("Antal döda per dag")

def grafer(x,y):
  fig, axs = plt.subplots(1, 2, figsize=(16, 6))
  axs[0].plot(y)
  axs[0].set_xlabel("Dagen")
  axs[0].set_ylabel("Totala döda för dagen")
  fig.suptitle('Människor som har dött per dag/månad')
grafer(x,döda)
```
