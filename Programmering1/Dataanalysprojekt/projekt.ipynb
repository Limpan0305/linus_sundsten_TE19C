```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Gender_Data.csv") #Läser in filen med hjälp av pandas
smittade = df["Total_Cases"]
färger = ["blue", "pink"]
döda = df["Total_Deaths"]
x =("Män", "Kvinnor")

def grafer(x,y):
  fig, axs = plt.subplots(1, 2, figsize=(16, 6))
  axs[0].bar(x, y, width=0.8, color=färger, align="center")
  axs[0].set_xlabel("Kön")
  axs[0].set_ylabel("Totala fall")
  benämningar = 'Manliga fall', 'Kvinnliga fall'
  axs[1].pie(y, labels = benämningar, autopct = '%1.1f%%', colors=färger) 
  fig.suptitle('Manliga & Kvinnliga totala fall och dödsfall')
grafer(x,döda)
```

```python
import plotly.express as px
import numpy as np
import pandas as pd
antal_kast = 100

#simulera tärningskast
tärningar = np.random.randint(1,7,antal_kast)

#räkna antalet ettor, tvåor, treor, fyror ...
utfall = np.bincount(tärningar)

#skapa en dictionary
frekvens_dict = dict(sida=np.arange(1,7), utfall = utfall)

#skapa en dataframe från dictionary
df = pd.DataFrame(frekvens_dict )

#plotly express för att rita ett stapeldiagram
fig = px.bar(df, x="sida", y="Antal", title=f"Antal kast med en tärning {antal_kast} st")
fig.show()

print(frekvens_dict)
print(f"Tärningskast: {tärningar}")
print(f"Utfall per sida: {utfall}")
print(df)
```
