```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Gender_Data.csv") #Lser in filen med hjlp av pandas
smittade = df["Total_Cases"]
frger = ["blue", "pink"]
dda = df["Total_Deaths"]
x =("Mn", "Kvinnor")

def grafer(x,y):
  fig, axs = plt.subplots(1, 2, figsize=(16, 6))
  axs[0].bar(x, y, width=0.8, color=frger, align="center")
  axs[0].set_xlabel("Kn")
  axs[0].set_ylabel("Totala fall")
  benmningar = 'Manliga fall', 'Kvinnliga fall'
  axs[1].pie(y, labels = benmningar, autopct = '%1.1f%%', colors=frger) 
  fig.suptitle('Manliga & Kvinnliga totala fall och ddsfall')
grafer(x,dda)
```

```python
import plotly.express as px
import numpy as np
import pandas as pd
antal_kast = 100

#simulera trningskast
trningar = np.random.randint(1,7,antal_kast)

#rkna antalet ettor, tvor, treor, fyror ...
utfall = np.bincount(trningar)

#skapa en dictionary
frekvens_dict = dict(sida=np.arange(1,7), utfall = utfall)

#skapa en dataframe frn dictionary
df = pd.DataFrame(frekvens_dict )

#plotly express fr att rita ett stapeldiagram
fig = px.bar(df, x="sida", y="Antal", title=f"Antal kast med en trning {antal_kast} st")
fig.show()

print(frekvens_dict)
print(f"Trningskast: {trningar}")
print(f"Utfall per sida: {utfall}")
print(df)
```
